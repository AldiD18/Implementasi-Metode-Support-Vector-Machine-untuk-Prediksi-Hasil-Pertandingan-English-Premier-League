```python
# Membuat fungsi menghitung selisih rating Home dengan Away
def get_rating_diff(season):
    season['DiffOvrPES'] = season['HOvrPES'] - season['AOvrPES']
    season['DiffDefPES'] = season['HDefPES'] - season['ADefPES']
    season['DiffMidPES'] = season['HMidPES'] - season['AMidPES']
    season['DiffFwdPES'] = season['HFwdPES'] - season['AFwdPES']
    season['DiffPhyPES'] = season['HPhyPES'] - season['APhyPES']
    season['DiffSpdPES'] = season['HSpdPES'] - season['ASpdPES']
    
    # Normalisasi Rating PES dengan/berdasarkan nilai maksimumnya, untuk mendapatkan rentang nilai 0 sampai 1 
    season['HOvrPES'] = season['HOvrPES'] / season['HOvrPES'].max()
    season['AOvrPES'] = season['AOvrPES'] / season['AOvrPES'].max()
    season['HDefPES'] = season['HDefPES'] / season['HDefPES'].max()
    season['ADefPES'] = season['ADefPES'] / season['ADefPES'].max()
    season['HMidPES'] = season['HMidPES'] / season['HMidPES'].max()
    season['AMidPES'] = season['AMidPES'] / season['AMidPES'].max()
    season['HFwdPES'] = season['HFwdPES'] / season['HFwdPES'].max()
    season['AFwdPES'] = season['AFwdPES'] / season['AFwdPES'].max()
    season['HPhyPES'] = season['HPhyPES'] / season['HPhyPES'].max()
    season['APhyPES'] = season['APhyPES'] / season['APhyPES'].max()
    season['HSpdPES'] = season['HSpdPES'] / season['HSpdPES'].max()
    season['ASpdPES'] = season['ASpdPES'] / season['ASpdPES'].max()
    season['DiffOvrPES'] = season['DiffOvrPES'] / season['DiffOvrPES'].max()
    season['DiffDefPES'] = season['DiffDefPES'] / season['DiffDefPES'].max()
    season['DiffMidPES'] = season['DiffMidPES'] / season['DiffMidPES'].max()
    season['DiffFwdPES'] = season['DiffFwdPES'] / season['DiffFwdPES'].max()
    season['DiffPhyPES'] = season['DiffPhyPES'] / season['DiffPhyPES'].max()
    season['DiffSpdPES'] = season['DiffSpdPES'] / season['DiffSpdPES'].max()
    
    return season


# Terapkan fungsi
final_1011 = get_rating_diff(final_1011)             
final_1112 = get_rating_diff(final_1112) 
final_1213 = get_rating_diff(final_1213) 
final_1314 = get_rating_diff(final_1314) 
final_1415 = get_rating_diff(final_1415) 
final_1516 = get_rating_diff(final_1516) 
final_1617 = get_rating_diff(final_1617) 
final_1718 = get_rating_diff(final_1718) 
final_1819 = get_rating_diff(final_1819) 
```

2. Akumulasi Goal Difference "Home & Away" (untuk set Fitur A) 
    - serta Diferensial/selisih Goal Difference Home dengan Away (untuk set Fitur B)
    - dibuat terlebih dahulu Fungsi Gol Dicetak dan Kemasukannya (dikurangi), untuk membuat Goal Difference 

*Akumulasi sebelum ke pertandingan baru, tiap matchweek/minggu pertandingan & tiap musimnya*

```python
# 1. Membuat fungsi akumulasi gol dicetak tim
def get_compute_goals_scored(season):
    # Buat dictionary/kamus dengan nama tim sebagai key
    teams = {}
    for i in season.groupby('HomeTeam').mean().T.columns:  # Nama tim sebagai indeks
        teams[i] = []
    
    # Gol dicetak setiap matchweek oleh masing-masing tim (Home atau Away)
    for i in range(len(season)):
        HTGSco = season.iloc[i]['FTHG']
        ATGSco = season.iloc[i]['FTAG']
        teams[season.iloc[i].HomeTeam].append(HTGSco)
        teams[season.iloc[i].AwayTeam].append(ATGSco)
        
    # Buat dataframe untuk gol yang dicetak di mana baris adalah tim dan kolom adalah matchweek
    goals_scored = pd.DataFrame(data=teams, index=[i for i in range(1,(len(season) // 10) + 1)]).T
    goals_scored[0] = 0
    
    # Akumulasi gol dicetak setiap matchweek
    for i in range(2,(len(season) // 10) + 1):
        goals_scored[i] = goals_scored[i] + goals_scored[i-1]
        
    return goals_scored


# 2. Membuat fungsi akumulasi gol kemasukan tim
def get_compute_goals_conceded(season):
    # Buat dictionary/kamus dengan nama tim sebagai key
    teams = {}
    for i in season.groupby('HomeTeam').mean().T.columns:  
        teams[i] = []
    
    # Gol kemasukan setiap matchweek oleh masing-masing tim (Home atau Away)
    for i in range(len(season)):
        ATGConc = season.iloc[i]['FTHG']
        HTGConc = season.iloc[i]['FTAG']
        teams[season.iloc[i].HomeTeam].append(HTGConc)
        teams[season.iloc[i].AwayTeam].append(ATGConc)
        
    # Buat dataframe untuk gol kemasukan di mana baris adalah tim dan kolom adalah matchweek    
    goals_conceded = pd.DataFrame(data=teams, index=[i for i in range(1,(len(season) // 10) + 1)]).T
    goals_conceded[0] = 0
    
    # Akumulasi gol kemasukan setiap matchweek
    for i in range(2,(len(season) // 10) + 1):
        goals_conceded[i] = goals_conceded[i] + goals_conceded[i-1]
        
    return goals_conceded


# 3. Penggabungan fungsi gol dicetak dan kemasukan (pembuatan Goal Difference)
def get_combined(season):
    AccuGSco = get_compute_goals_scored(season)
    AccuGConc = get_compute_goals_conceded(season)
    
    j = 0
    HTGSco = []
    ATGSco = []
    HTGConc = []
    ATGConc = []
    
    for i in range(len(season)):
        home_team = season.iloc[i].HomeTeam
        away_team = season.iloc[i].AwayTeam
        HTGSco.append(AccuGSco.loc[home_team][j])
        ATGSco.append(AccuGSco.loc[away_team][j])
        HTGConc.append(AccuGConc.loc[home_team][j])
        ATGConc.append(AccuGConc.loc[away_team][j])
        
        if ((i + 1) % 10) == 0:
            j = j + 1
            
    season['HTGSco'] = HTGSco
    season['ATGSco'] = ATGSco
    season['HTGConc'] = HTGConc
    season['ATGConc'] = ATGConc
    
    season['HTGDference'] = season['HTGSco'] - season['HTGConc']  # Goal Difference (selisih gol dicetak dan kemasukan)
    season['ATGDference'] = season['ATGSco'] - season['ATGConc']
    season['DiffGDference'] = season['HTGDference'] - season['ATGDference']
    
    season.drop(['HTGSco','ATGSco','HTGConc','ATGConc'],inplace=True,axis=1,errors='ignore') #yg dibutuhkan goal difference saja
    
    return season


# Terapkan fungsi ke tiap musim
final_1011 = get_combined(final_1011)                      
final_1112 = get_combined(final_1112)
final_1213 = get_combined(final_1213)
final_1314 = get_combined(final_1314)
final_1415 = get_combined(final_1415)
final_1516 = get_combined(final_1516)
final_1617 = get_combined(final_1617)
final_1718 = get_combined(final_1718)
final_1819 = get_combined(final_1819)
```

3. Menambahkan (l) pertandingan sebelumnya untuk statistik tim "Home & Away" yang mencakup: (untuk set Fitur A)
    - Goals, karena menentukan hasil akhir pertandingan.
    - Shots, karena asumsi bahwa semakin banyak tendangan akan menghasilkan banyak gol.
    - ShotsOnTarget, karena menentukan banyaknya tendangan langsung ke arah gawang memungkinkan terjadinya gol.
    - Corners, karena menentukan tim bermain dengan baik dengan memaksa tendangan sepak pojok dari lawan.
        
        -(serta Diferensial/selisihnya untuk set Fitur B)

```python
# Membuat fungsi yang menghitung l pertandingan sebelumnya untuk statistik tim 
def compute_past_perform(season, slidingWindowParameter):
    
    # Mengatur Parameter (nilai pertandingan sebelumnya) ke l
    l = slidingWindowParameter

    # Membuat list/daftar semua tim yang bermain pada musimnya
    Teams = list((season).HomeTeam.unique())
    
    # Inisialisasi nilai awal kolom (dengan nilai kosong)
    season['HGLPP'] = np.nan
    season['AGLPP'] = np.nan
    season['HSLPP'] = np.nan
    season['ASLPP'] = np.nan
    season['HSTLPP'] = np.nan
    season['ASTLPP'] = np.nan
    season['HCLPP'] = np.nan
    season['ACLPP'] = np.nan
    
    # Membuat DataFrame Sementara yang terdiri dari catatan tim bertanding
    for z in range(0, 20):

        # Membuat DataFrame Sementara tempat tim berada "Home" atau "Away"
        tempDataFrame = season[(season['HomeTeam'] == str(Teams[z])) | ( season['AwayTeam'] == str(Teams[z]))]

        # Membuat list/daftar untuk tim yang diamati
        Goals = []
        Shots = []
        ShotsOnTarget = []
        Corners = []

        for index, row in tempDataFrame.iterrows():
    
            if (Teams[z] == row['HomeTeam']):
                Goals.append(float(row['FTHG']))
                Shots.append(float(row['HS']))
                ShotsOnTarget.append(float(row['HST']))
                Corners.append(float(row['HC']))

            elif (Teams[z] == row['AwayTeam']):
                Goals.append(float(row['FTAG']))
                Shots.append(float(row['HS']))
                ShotsOnTarget.append(float(row['AST']))
                Corners.append(float(row['AC']))
                
        # Membuat daftar untuk menyimpan nilai untuk Fitur LPP yang sesuai
        # Karena fitur-fitur ini tidak ada untuk pertandingan l pertama dari setiap tim, isikan NAN untuk nilai l pertama
        Goals_LPP = [np.nan] * l
        Shots_LPP = [np.nan] * l
        ShotsOnTarget_LPP = [np.nan] * l
        Corners_LPP = [np.nan] * l
        
        # Menambahkan nilai yang sesuai ke List/Daftar
        # Jumlah perhitungan yang dilakukan adalah (n + 1 - l) di mana:
        ## n = jumlah pertandingan dalam musim untuk masing-masing tim (38)
        ## l = sliding window hyper-parameter
        for i in range(0, (39 - l)):

            # Memperoleh jumlah irisan untuk diamati
            # Jumlahkan irisan dan "Normalisasikan dengan l"
            Goal_SliceSum = sum(Goals[i : (i + l)])/l
            Shot_SliceSum = sum(Shots[i : (i + l)])/l
            ShotsOnTarget_SliceSum = sum(ShotsOnTarget[i : (i + l)])/l
            Corner_SliceSum = sum(Corners[i : (i + l)])/l

            # Menambahkan ke list/daftar fitur LPP yang sesuai
            Goals_LPP.append(Goal_SliceSum)
            Shots_LPP.append(Shot_SliceSum)
            ShotsOnTarget_LPP.append(ShotsOnTarget_SliceSum)
            Corners_LPP.append(Corner_SliceSum)
            
        # Membuat list/daftar untuk nilai indeks dari game yang terkandung dalam tempDataFrame
        game_Index = tempDataFrame.index.tolist()

        # Membuat dua list/daftar yang berisi nomor indeks dari game-game di mana tim yang diamati adalah Home atau Away
        index_Home = []
        index_Away = []

        # Pisahkan indeks pertandingan kandang dan tandang
        for index, row in tempDataFrame.iterrows():
            
            if (Teams[z] == row['HomeTeam']):
                 index_Home.append(index)

            elif (Teams[z] == row['AwayTeam']):
                index_Away.append(index)

        # Menambahkan nilai "LPP" yang sesuai ke dataframe
        for j in range(0, 38):

            if (game_Index[j] in index_Home):
                season['HGLPP'][game_Index[j]] = Goals_LPP[j]
                season['HSLPP'][game_Index[j]] = Shots_LPP[j]
                season['HSTLPP'][game_Index[j]] = ShotsOnTarget_LPP[j]
                season['HCLPP'][game_Index[j]] = Corners_LPP[j]

            elif (game_Index[j] in index_Away):
                season['AGLPP'][game_Index[j]] = Goals_LPP[j]
                season['ASLPP'][game_Index[j]] = Shots_LPP[j] 
                season['ASTLPP'][game_Index[j]] = ShotsOnTarget_LPP[j]
                season['ACLPP'][game_Index[j]] = Corners_LPP[j]
        
    
    # Isi kolom dengan hasil nilainya
    season['DiffGLPP'] = season.apply(lambda row: row['HGLPP'] - row['AGLPP'], axis = 1)
    season['DiffSLPP'] = season.apply(lambda row: row['HSLPP'] - row['ASLPP'], axis = 1)
    season['DiffSTLPP'] = season.apply(lambda row: row['HSTLPP'] - row['ASTLPP'], axis = 1)
    season['DiffCLPP'] = season.apply(lambda row: row['HCLPP'] - row['ACLPP'], axis = 1)
    
    return season


# Terapkan fungsi ke tiap musim
final_1011 = compute_past_perform(final_1011, 6)                      
final_1112 = compute_past_perform(final_1112, 6)  
final_1213 = compute_past_perform(final_1213, 6)  
final_1314 = compute_past_perform(final_1314, 6)  # Mengatur nilai l adalah 6
final_1415 = compute_past_perform(final_1415, 6)  
final_1516 = compute_past_perform(final_1516, 6)  
final_1617 = compute_past_perform(final_1617, 6)  
final_1718 = compute_past_perform(final_1718, 6)  
final_1819 = compute_past_perform(final_1819, 6)  
```

4. Menambahkan Streak/perolehan poin (l) pertandingan sebelumnya "Home & Away" (untuk set Fitur A)

    -serta Diferensial Streak (untuk set Fitur B)

```python
# Membuat fungsi yang menghitung Streak
def compute_streak(season, slidingWindowParameter):
    
    # Mengatur Parameter (nilai performa sebelumnya) ke l
    l = slidingWindowParameter

    # Membuat list/daftar semua tim yang bermain pada musimnya
    Teams = list((season).HomeTeam.unique())
    
    # Inisialisasi nilai awal kolom (dengan nilai kosong)
    season['HStreak'] = np.nan
    season['AStreak'] = np.nan
    
    # Membuat DataFrame Sementara yang terdiri dari catatan pertandingan tim
    for z in range(0, 20):

        # Membuat DataFrame Sementara tempat tim berada "Home" atau "Away"
        tempDataFrame = season[(season['HomeTeam'] == str(Teams[z])) | ( season['AwayTeam'] == str(Teams[z]))]
    
        # Membuat list/daftar yang berisi poin yang didapat untuk masing-masing tim setelah pertandingan
        ## 0 - Kalah
        ## 1 - Imbang
        ## 3 - Menang
        match_points = []

        # Membuat list/daftar yang berisi bobot yang ditetapkan untuk setiap pertandingan 
        for index , row in tempDataFrame.iterrows():
            
            if (Teams[z] == row['HomeTeam']):
                if (row['FTR'] == 'A') :
                    match_points.append(0.0)
                elif (row['FTR'] == 'D') :
                    match_points.append(1.0)
                elif (row['FTR'] == 'H') :
                    match_points.append(3.0)

            elif (Teams[z] == row['AwayTeam']):
                if (row['FTR'] == 'H') :
                    match_points.append(0.0)
                elif (row['FTR'] == 'D') :
                    match_points.append(1.0)
                elif (row['FTR'] == 'A') :
                    match_points.append(3.0)
        
        # Membuat list/daftar untuk menyimpan nilai untuk Fitur Streak yang sesuai
        # Karena fitur tidak ada untuk pertandingan l pertama dari setiap tim, isikan NAN untuk nilai l pertama
        streak = [np.nan] * l
        
        # Menambahkan nilai yang sesuai ke List/Daftar
        # Jumlah perhitungan yang dilakukan adalah (n + 1 - l) di mana:
        ## n = jumlah pertandingan dalam musim untuk masing-masing tim (38)
        ## l = sliding window hyper-parameter
        for i in range(0, (39 - l)):

            # Memperoleh jumlah irisan untuk diamati
            get_PointsSlice = match_points[i : (i + l)]

            # Jumlahkan irisan dan "Normalisasikan dengan 3l untuk mendapatkan rentang nilai 0 sampai 1"
            streak_value = sum(get_PointsSlice)/(3 * l)

            # Menambah list/daftar fitur yang sesuai
            streak.append(streak_value)
            
        # Membuat list/daftar untuk nilai indeks dari game yang terkandung dalam tempDataFrame
        game_Index = tempDataFrame.index.tolist()

        # Membuat dua list/daftar yang berisi nomor indeks dari game-game di mana tim yang diamati adalah Home atau Away
        index_Home = []
        index_Away = []

        # Pisahkan indeks pertandingan kandang dan tandang
        for index, row in tempDataFrame.iterrows():

            if (Teams[z] == row['HomeTeam']):
                 index_Home.append(index)

            elif (Teams[z] == row['AwayTeam']):
                index_Away.append(index)

        # Menambahkan nilai "Streak" yang sesuai ke dataframe
        for j in range(0, 38):

            if (game_Index[j] in index_Home):
                season['HStreak'][game_Index[j]] = streak[j]

            elif (game_Index[j] in index_Away):
                season['AStreak'][game_Index[j]] = streak[j]
                
    
    # Isi kolom dengan hasil nilainya
    season['DiffStreak'] = season.apply(lambda row: row['HStreak'] - row['AStreak'], axis = 1)
    
    return season


# Terapkan fungsi ke tiap musim
final_1011 = compute_streak(final_1011, 6)                      
final_1112 = compute_streak(final_1112, 6)  
final_1213 = compute_streak(final_1213, 6)  
final_1314 = compute_streak(final_1314, 6)  # Mengatur nilai l adalah 6 
final_1415 = compute_streak(final_1415, 6)  
final_1516 = compute_streak(final_1516, 6)  
final_1617 = compute_streak(final_1617, 6)  
final_1718 = compute_streak(final_1718, 6)  
final_1819 = compute_streak(final_1819, 6)
```

5. Akumulasi poin/pts "Home & Away" (untuk set Fitur A)

    -serta Diferensial Pts (untuk set Fitur B)

*Akumulasi sebelum ke pertandingan baru, tiap matchweek/minggu pertandingan & tiap musimnya*

```python
# Identifikasi poin yang didapat
def get_points(result):
    if result == 'W':
        return 3.0  # Menang 3 poin
    elif result == 'D':
        return 1.0  # Imbang 1 poin
    else:
        return 0.0  # Kalah 0 poin

# Fungsi menambahkan poin    
def get_add_points(match_results, season):
    matchres_points = match_results.applymap(get_points)
    for i in range(2, (len(season) // 10) + 1):
        matchres_points[i] = matchres_points[i] + matchres_points[i-1]
        
    matchres_points.insert(column=0, loc=0, value=[0 * i for i in range(20)])
    return matchres_points

# Fungsi konversi poin yang didapat sesuai "identifikasi poin"
def get_match_results(season):
    # Buat dictionary/kamus dengan nama tim sebagai key
    teams = {}
    for i in season.groupby('HomeTeam').mean().T.columns:
        teams[i] = []
        
    # Value/nilai yang terkait dengan key/kunci adalah list/daftar yang berisi hasil pertandingan
    for i in range(len(season)):
        if season.iloc[i].FTR == 'H':
            teams[season.iloc[i].HomeTeam].append('W')
            teams[season.iloc[i].AwayTeam].append('L')
        elif season.iloc[i].FTR == 'A':
            teams[season.iloc[i].HomeTeam].append('L')
            teams[season.iloc[i].AwayTeam].append('W')
        else:
            teams[season.iloc[i].HomeTeam].append('D')
            teams[season.iloc[i].AwayTeam].append('D')
            
    return pd.DataFrame(data=teams, index=[i for i in range(1, (len(season) // 10) + 1)]).T


# Fungsi akumulasi poinnya
def get_accu_points(season):
    match_results = get_match_results(season)
    add_pts = get_add_points(match_results, season)
    HTPts = []
    ATPts = []
    j = 0
    for i in range(len(season)):
        ht = season.iloc[i].HomeTeam
        at = season.iloc[i].AwayTeam
        HTPts.append(add_pts.loc[ht][j])
        ATPts.append(add_pts.loc[at][j])
        
        if ((i + 1) % 10) == 0:
            j = j + 1
            
    season.loc[:,'HTPts'] = HTPts
    season.loc[:,'ATPts'] = ATPts
    
    season['DiffPts'] = season['HTPts'] - season['ATPts']
    
    return season


# Terapkan fungsi ke tiap musim
final_1011 = get_accu_points(final_1011)                      
final_1112 = get_accu_points(final_1112)
final_1213 = get_accu_points(final_1213)
final_1314 = get_accu_points(final_1314)
final_1415 = get_accu_points(final_1415)
final_1516 = get_accu_points(final_1516)
final_1617 = get_accu_points(final_1617)
final_1718 = get_accu_points(final_1718)
final_1819 = get_accu_points(final_1819)
```

6. Menambahkan kolom Matchweek (Minggu Pertandingan) sebagai Identifikasi 

    -serta digunakan untuk Normalisasi fitur Pts

```python
# Fungsi "matchweek", yaitu tiap 10 baris adalah 1 matchweek
def get_mw(season):
    j = 1
    match_week = []
    for i in range(len(season)):
        match_week.append(j)
        if ((i + 1) % 10) == 0:
            j = j + 1
    season['MW'] = match_week
    return season

# Terapkan fungsi ke tiap musim
final_1011 = get_mw(final_1011)                      
final_1112 = get_mw(final_1112)
final_1213 = get_mw(final_1213)
final_1314 = get_mw(final_1314)
final_1415 = get_mw(final_1415)
final_1516 = get_mw(final_1516)
final_1617 = get_mw(final_1617)
final_1718 = get_mw(final_1718)
final_1819 = get_mw(final_1819)
```

# finalisasi dataset

- penggabungan (per musim menjadi 1 data seluruh musim)

```python
# Dataframe tiap musim digabung menjadi satu dataframe

final_data = pd.concat([final_1011,final_1112,final_1213,final_1314,final_1415,final_1516,final_1617,
                        final_1718,final_1819], ignore_index=True, sort=True)
final_data = final_data.drop(['Unnamed: 0'], axis=1)
```

- normalisasi fitur ke Matchweek

```python
# Normalisasi fitur Pts (yg blm dinormalisasi sebelumnya) untuk meminimalkan range nilainya
# fitur Pts menghasilkan rentang nilai 0 sampai 3

cols = ['HTPts', 'ATPts', 'DiffPts']

for col in cols:
    final_data[col] = final_data[col] / final_data.MW 
```
